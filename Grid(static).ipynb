```python
# Draw a static grid (True)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o')
plt.xlabel("X values (Independent Variable)")
plt.ylabel("Y values (Dependent Variable)")
plt.title("X-Y Axis Data Plot")

plt.grid(True)
plt.show()
```

```python

# Draw a static grid (False)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.grid(False)
plt.title("Static Grid = False")

plt.show()
```
